```python
import numpy as np
import matplotlib.pyplot as plt

# Dados experimentais
execution_times_data = {
    "10x10": 0.0203462,
    "20x20": 0.34115,
    "30x30": 1.83185,
    "40x40": 5.49026
}

# Extrao de valores de n e tempos
sizes = [(int(size.split('x')[0]))**2 for size in execution_times_data.keys()]
times = list(execution_times_data.values())

# Clculo terico: O(n^2 log n)
theoretical_growth = [n**2 * np.log2(n^2) for n in sizes]

# Normalizando a funo terica para comparar com os tempos experimentais
scale_factor = times[0] / theoretical_growth[0]  # Normaliza para o primeiro ponto experimental
scaled_theoretical_growth = [scale_factor * value for value in theoretical_growth]

# Plotando os dados experimentais e tericos
plt.figure(figsize=(10, 6))
plt.plot(sizes, times, marker='o', linestyle='-', label='Tempos Experimentais')
plt.plot(sizes, scaled_theoretical_growth, marker='x', linestyle='--', label=r'$O(V^2 \log V)$ (Normalizado)')
plt.title('Comparao entre Tempos - Encontrar as estaes em cada regio', fontsize=14)
plt.xlabel('Tamanho do Grid $(n^2 = V)$', fontsize=12)
plt.ylabel('Tempo de Execuo (segundos)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados experimentais
execution_times_data = {
    # "10x10": 0.00039179,
    # "20x20": 0.00331083,
    # "30x30": 0.0104651
    
    "10x10": 0.000377707,
    "20x20": 0.00301607,
    "30x30": 0.0104101
}

# Extrao de valores de n e tempos
sizes = [int(size.split('x')[0])**2 for size in execution_times_data.keys()]
times = list(execution_times_data.values())


# n^2 = V
# O(T^2 ((V+E)logV) + ElogE + VlogV), T = n/2
# O((n/2)^2 ((V)logV))
# O((V^2/ 2)  VlogV)
# O((n/2)^2 (n^2log(n^2)))
# Clculo terico: O((n/2)^2 (n^2log(n^2)))
# theoretical_growth = [ (n**4 * np.log2(n**2)) for n in sizes]

theoretical_growth = [ ((n**3 * np.log2(n**2))) for n in sizes]

# Normalizando a funo terica para comparar com os tempos experimentais
scale_factor = times[0] / theoretical_growth[0]  # Normaliza para o primeiro ponto experimental
scaled_theoretical_growth = [scale_factor * value for value in theoretical_growth]

# Plotando os dados experimentais e tericos
plt.figure(figsize=(10, 6))
plt.plot(sizes, times, marker='o', linestyle='-', label='Tempos Experimentais')
plt.plot(sizes, scaled_theoretical_growth, marker='x', linestyle='--', label=r'$O(n^3 \log n^2)$ (Normalizado)')
plt.title('Comparao entre Tempos - Encontrar as linhas do metro', fontsize=14)
plt.xlabel('Tamanho do Grid $(n^2 = V)$', fontsize=12)
plt.ylabel('Tempo de Execuo (segundos)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
