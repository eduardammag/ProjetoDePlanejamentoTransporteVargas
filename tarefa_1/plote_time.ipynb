```python
import numpy as np
import matplotlib.pyplot as plt

# Dados experimentais
execution_times_data = {
    "10x10": 0.0203462,
    "20x20": 0.34115,
    "30x30": 1.83185,
    "40x40": 5.49026
}

# Extração de valores de n e tempos
sizes = [int(size.split('x')[0])**2 for size in execution_times_data.keys()]  # V = n^2
times = list(execution_times_data.values())

# Cálculo teórico: O(V^2 log V), onde V = n^2
theoretical_growth = [n**2 * np.log2(n**2) for n in sizes]

# Normalizando a função teórica para comparar com os tempos experimentais
scale_factor = times[0] / theoretical_growth[0]  # Normaliza pelo primeiro ponto experimental
scaled_theoretical_growth = [scale_factor * value for value in theoretical_growth]

# Definindo constantes c1 e c2 para os limites superior e inferior
c1, c2 = 0.5, 1  # Ajuste conforme necessário para verificar os limites

# Calculando limites inferior e superior
lower_bound = [c1 * value for value in scaled_theoretical_growth]
upper_bound = [c2 * value for value in scaled_theoretical_growth]

# Plotando os dados experimentais, teóricos e os limites
plt.figure(figsize=(10, 6))
plt.plot(sizes, times, marker='o', linestyle='-', label='Tempos Experimentais', linewidth=2)
# plt.plot(sizes, scaled_theoretical_growth, marker='x', linestyle='--', label=r'$O(V^2 \log V)$ (Normalizado)', linewidth=2)
plt.plot(sizes, lower_bound, linestyle=':', label=r'$c_1 \cdot O(V^2 \log V)$', linewidth=2)
plt.plot(sizes, upper_bound, linestyle=':', label=r'$c_2 \cdot O(V^2 \log V)$', linewidth=2)
plt.title('Comparação entre Tempos - Encontrar as estações em cada região', fontsize=14)
plt.xlabel('Tamanho do Grid $(n^2 = V)$', fontsize=12)
plt.ylabel('Tempo de Execução (segundos)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados experimentais
execution_times_data = {
    "10x10": 0.000377707,
    "20x20": 0.00301607,
    "30x30": 0.0104101
}

# Extração de valores de n e tempos
sizes = [int(size.split('x')[0])**2 for size in execution_times_data.keys()]  # V = n^2
times = list(execution_times_data.values())


# n^2 = V
# O(T^2 ((V+E)logV) + ElogE + VlogV), T = n/2
# O((n/2)^2 ((V)logV))
# O((V^2/ 2)  VlogV)
# O((n/2)^2 (n^2log(n^2)))
# Cálculo teórico: O((n/2)^2 (n^2log(n^2)))
# Função teórica de crescimento: O(n^3 log(n^2))
theoretical_growth = [n**3 * np.log2(n**2) for n in sizes]

# Normalizando a função teórica
scale_factor = times[0] / theoretical_growth[0]  # Normalizar pelo primeiro ponto experimental
scaled_theoretical_growth = [scale_factor * value for value in theoretical_growth]

# Definindo constantes c1 e c2
c1, c2 = 0.01, 0.09  # Constantes ajustáveis para os limites superior e inferior

# Calculando limites inferior e superior
lower_bound = [c1 * value for value in scaled_theoretical_growth]
upper_bound = [c2 * value for value in scaled_theoretical_growth]

# Plotando os dados experimentais, teóricos e os limites
plt.figure(figsize=(10, 6))
plt.plot(sizes, times, marker='o', linestyle='-', label='Tempos Experimentais', linewidth=2)
# plt.plot(sizes, scaled_theoretical_growth, marker='x', linestyle='--', label=r'$O(n^3 \log n^2)$ (Normalizado)', linewidth=2)
plt.plot(sizes, lower_bound, linestyle=':', label=r'$c_1 \cdot O(V^{3/2} \log V)$', linewidth=2)
plt.plot(sizes, upper_bound, linestyle=':', label=r'$c_2 \cdot O(V^{3/2} \log V)$', linewidth=2)
plt.title('Comparação entre Tempos Experimentais e Limites Teóricos', fontsize=14)
plt.xlabel('Tamanho do Grid $(n^2 = V)$', fontsize=12)
plt.ylabel('Tempo de Execução (segundos)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
